# Movie Recommender System

```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
# Merge the datasets on 'Title'

movies=movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
# Picking the relvant data columns for our movie recommender system

movies=movies[['id','genres','keywords','title','overview','cast','crew']]
```

```python
movies.head()
```

```python
movies.shape
```

```python
movies.isnull().sum()
```

```python
# we have pretty good size of data so we can drop three null values of overview
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
# To check wether a duplicate value is present or not

movies.duplicated().sum()
```

```python
movies.genres[0]
```

```python
import ast
ast.literal_eval(movies.genres[0])[1]['name']
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies.genres.apply(convert)
```

```python
movies.head()
```

```python
# applying convert function on 'keywords' column

movies['keywords']=movies.keywords.apply(convert)
```

```python
movies.head()
```

```python
# applying convert function on 'cast' column
movies['cast']=movies.cast.apply(convert)
```

```python
movies.head()
```

```python
# Fetching directors name from the column crew

def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
    return l
```

```python
movies['crew']=movies.crew.apply(fetch_director)
```

```python
movies.head()
```

```python
movies.info()
```

```python
#Movie['overview'] was the only column which was in string format . so we converted that into list so that we can appeand 
#that column with every other list categorical column 


movies['overview']=movies['overview'].apply(lambda x : x.split())
```

```python
movies.overview[0]
```

```python
movies.cast[0][1].replace(" ","")
```

```python
movies['cast']=movies['cast'].apply(lambda x :[i.replace(" ","") for i in x])
movies['genres']=movies['genres'].apply(lambda x :[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x :[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x :[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# Concat Keywords genres, overview ,cast and crew columns

movies['tags']=movies['genres']+movies['keywords']+movies['overview']+movies['cast']+movies['crew']
```

```python
movies.tags[0]
```

```python
# convert list into string

movies['tags']=movies.tags.apply(lambda x:" ".join(x))
```

```python
movies.head()
```

```python
# creating a new dataset and dropping irreleavent values

movies_updated=movies[['id','title','tags']]
```

```python
movies_updated.head()
```

```python
movies_updated['tags'][0].split()
```

```python
from nltk.stem.porter import PorterStemmer

ps=PorterStemmer()
```

```python
# Example

ls=['loving','loved','love']

l=[]
for i in ls:
    l.append(ps.stem(i))
print(l)
```

 Basically stem converts a word in to it's root form
 
 let's apply this to our tags column so similar words will get converted into root form

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
        
    return " ".join(y)
```

```python
movies_updated['tags']=movies_updated['tags'].apply(stem)
```

```python
# count Vectorizer on tags column

from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
Vectors=cv.fit_transform(movies_updated['tags']).toarray()
```

```python
Vectors[0].shape
```

```python
# Most frequent words choose by CountVectorizer

cv.get_feature_names()
```

```python
# Cosine similarity have value between 0 and 1

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(Vectors)
```

```python
similarity[0]
```

```python
similarity[0].shape
```

So I have calculated the distance of first movie with every other movies

```python
index_movie=movies_updated[movies_updated.title=="Avatar"].index[0]
```

```python
sorted(list(enumerate(similarity[index_movie])),reverse=True,key=lambda x : x[1])[1:6]
```

```python
movies_updated.iloc[582].title
```

```python
# Now i need to create a funtion which will fetch the movies having closest distance  from the input movie

def recommend(movie):
    movies_index=movies_updated[movies_updated['title'] == movie].index[0]
    movie_distance=similarity[movies_index]
    movie_list=sorted(list(enumerate(movie_distance)),reverse=True,key=lambda x : x[1])[1:6]
    
    for i in movie_list:
        print(movies_updated.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

```python
import pandas as pd 
import numpy as np
```

```python
match=pd.read_csv('matches.csv')
```

```python
match.shape
```

```python
delivery=pd.read_csv('deliveries.csv')
```

```python
delivery.head()
```

```python
total_score=delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
total_score.head()
```

```python
total_score=total_score[total_score['inning']==1]
```

```python
match_df=match.merge(total_score[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
match_df.shape
```

```python
match_df['team1'].unique()
```

we need to replace names of the team with the current name of the team and we also need to remove few teams , who doesn't play 

```python
match_df['team1']=match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2']=match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1']=match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2']=match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df['team1'].unique()
```

```python
teams=[
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Delhi Capitals',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals'
]
```

```python
match_df=match_df[match_df['team1'].isin(teams)]
match_df=match_df[match_df['team2'].isin(teams)]
```

```python
# Now I have to take only those matches on which dl_applied is 0
match_df=match_df[match_df['dl_applied'] == 0]
```

```python
match_df.shape
```

```python
match_df=match_df[['match_id','city','winner','total_runs']]
```

```python
delivery_df=match_df.merge(delivery,on='match_id')
```

```python
delivery_df
```

```python
delivery_df=delivery_df[delivery_df['inning']==2]
```

```python
delivery_df['current_score']=delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
delivery_df['runs_left']=delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df
```

```python
delivery_df['balls_left']=126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df
```

```python
delivery_df['player_dismissed']=delivery_df['player_dismissed'].fillna('0').apply(lambda x: x if x=='0' else '1').astype(int)
```

```python
wickets=delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
```

```python
delivery_df['wickets_left']=10 - wickets
```

```python
delivery_df.groupby('match_id').get_group(114)
```

```python
delivery_df['batting_team']=delivery_df['batting_team'].replace('Deccan Chargers','Sunrisers Hyderabad')
delivery_df['bowling_team']=delivery_df['bowling_team'].replace('Deccan Chargers','Sunrisers Hyderabad')


delivery_df['batting_team']=delivery_df['batting_team'].replace('Delhi Daredevils','Delhi Capitals')
delivery_df['bowling_team']=delivery_df['bowling_team'].replace('Delhi Daredevils','Delhi Capitals')
```

```python
delivery_df
```

```python
# crr current run rate
# runs/over

delivery_df['crr']=delivery_df['current_score']*6/(120-delivery_df['balls_left'])
```

```python
# rrr 

delivery_df['rrr']=delivery_df['runs_left']*6/delivery_df['balls_left']
```

```python
delivery_df
```

```python

def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['results']=delivery_df.apply(result,axis=1)
```

```python
delivery_df
```

```python
final_df=delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets_left','total_runs_x','crr','rrr','results']]
```

```python
final_df.head()
```

```python
final_df=final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
final_df=final_df[final_df['balls_left']!=0]
```

```python
final_df.dropna(inplace=True)
```

# Model training

```python
from sklearn.model_selection import train_test_split

X=final_df.drop('results',axis=1)
y=final_df['results']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train.shape ,X_test.shape
```

Applying OneEndHotcoding on Columns with Categorical values

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(X_test)[5]
```
